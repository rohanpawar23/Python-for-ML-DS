# List Basics

### Things to remember About Lists:
* They use [] brackets and commas to separate objects in the
list. 
* Lists are mutable
* Lists support indexing and slicing.
* Lists can be nested.
* Lists also have a variety of useful methods that can be
called off of them.

```python
my_list = [1,2,3]
```

```python
type(my_list)
```

```python
my_list1 = [" Rohan."," I'm",["Hi",4]]
```

```python
my_list1[-1]
```

```python
my_list1[-1][0] + my_list1[1] + my_list1[0]
```

```python
print(my_list1[-1][0] + my_list1[1] + my_list1[0])
```

```python
len(my_list1)
```
